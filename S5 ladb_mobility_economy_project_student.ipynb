```python

import pandas as pd
traffic = pd.read_csv('/datasets/tomtom_traffic.csv')

# Convertir UpdateTimeUTC a datetime
traffic['UpdateTimeUTC'] = pd.to_datetime(traffic['UpdateTimeUTC'])

# Extraer el año en una nueva columna
traffic['year'] = traffic['UpdateTimeUTC'].dt.year

traffic_2024 = traffic[traffic['year'] == 2024]

# Calcular promedios
traffic_city_year_2024 = traffic_2024.groupby(['City','Country','year']).agg({
    'JamsDelay':'mean',
    'TrafficIndexLive':'mean',
    'JamsLengthInKms':'mean',
    'JamsCount':'mean',
    'MinsDelay':'mean',
    'TravelTimeLivePer10KmsMins':'mean',
    'TravelTimeHistoricPer10KmsMins':'mean'
}).reset_index()

# Mostrar resultado (corregido: sin salto de línea)
traffic_city_year_2024.head()
```

Resumen Ejecutivo

Relación entre movilidad urbana y productividad económica

Contexto y objetivo del análisis

La movilidad urbana se ha convertido en uno de los principales retos estructurales de las ciudades latinoamericanas, con impactos directos sobre la eficiencia económica, la calidad de vida y la competitividad regional. La congestión vehicular prolonga los tiempos de traslado, incrementa los costos operativos para empresas y trabajadores, y reduce la productividad efectiva del capital humano. En este contexto, el presente análisis tiene como objetivo principal examinar la relación entre la movilidad urbana —medida a través de indicadores de congestión y tiempos de viaje— y la productividad económica, representada por el Producto Interno Bruto (PIB) per cápita a nivel ciudad.

La pregunta central que guía este estudio es: ¿existe una relación observable entre altos niveles de congestión vehicular y menores indicadores de productividad económica en las principales ciudades de América Latina? Identificar esta relación es fundamental para orientar decisiones de política pública e inversión en infraestructura de transporte, especialmente en contextos urbanos donde los recursos son limitados y deben asignarse de forma estratégica.

Las variables clave utilizadas incluyen el Traffic Index Live y el tiempo promedio de viaje por cada 10 kilómetros, que funcionan como proxies de congestión urbana y eficiencia del sistema de transporte. Por otro lado, el PIB per cápita se emplea como un indicador sintético de productividad económica, ya que refleja la capacidad promedio de generación de valor económico por habitante. Estas variables son particularmente relevantes para tomadores de decisión, ya que conectan directamente la experiencia cotidiana de movilidad con resultados macroeconómicos observables.

Cobertura y alcance de los datos

El análisis se basa en un dataset consolidado que cubre los años 2023 y 2024, integrando información de movilidad urbana, desempeño económico y variables demográficas para un conjunto de ciudades representativas de América Latina. Entre las ciudades analizadas destacan Bogotá, Lima y Buenos Aires, seleccionadas por su tamaño poblacional, relevancia económica y notorios desafíos en materia de congestión vehicular.

A nivel geográfico, los datos incluyen múltiples países de la región, lo que permite realizar comparaciones entre ciudades bajo distintos contextos económicos y niveles de desarrollo. La estructura ciudad–año del dataset facilita el análisis longitudinal básico y la identificación de patrones temporales, aunque con ciertas limitaciones derivadas del corto horizonte temporal disponible.

Es importante destacar que, si bien el número de ciudades y países permite obtener una visión comparativa regional, el análisis no pretende establecer conclusiones causales definitivas, sino identificar patrones iniciales, contrastes relevantes y posibles outliers que justifiquen análisis más profundos o inversiones focalizadas.

Metodología (visión de alto nivel)

El proceso analítico comenzó con una limpieza y preparación exhaustiva de los datos, orientada a garantizar consistencia y comparabilidad entre fuentes. Esto incluyó la estandarización de formatos numéricos (por ejemplo, conversión de valores monetarios y eliminación de símbolos no numéricos), la normalización de nombres de columnas y la validación de tipos de datos para asegurar que las variables cuantitativas fueran correctamente interpretadas durante el análisis.

Posteriormente, los datos fueron agregados a nivel ciudad–año, permitiendo unificar la información de movilidad y economía bajo una misma unidad de análisis. Para integrar ambos conjuntos de datos se utilizó una unión INNER, asegurando que únicamente se consideraran observaciones con información completa y consistente tanto en indicadores de tráfico como en variables económicas. Esta decisión metodológica prioriza la calidad y coherencia de los datos por encima de la cobertura total.

Como parte de las validaciones exploratorias, se realizaron análisis visuales de distribuciones para identificar posibles valores atípicos, así como revisiones gráficas de tendencias generales entre congestión y PIB per cápita. Estas visualizaciones permitieron detectar patrones estructurales, estabilidad temporal en ciertos indicadores y posibles anomalías que podrían requerir una revisión adicional de fuentes o metodologías de medición.

Hallazgos principales

A nivel agregado y temporal, el análisis no muestra una correlación estadísticamente significativa fuerte entre los indicadores de congestión urbana y el PIB per cápita dentro del periodo 2023–2024. Este resultado se explica, en gran medida, por la baja variabilidad interanual de los índices de tráfico, los cuales se mantienen relativamente estables de un año a otro, mientras que el PIB per cápita presenta incrementos moderados en la mayoría de las ciudades analizadas.

No obstante, al analizar los datos desde una perspectiva comparativa entre ciudades, emergen patrones relevantes. En particular, Bogotá destaca consistentemente por presentar los niveles más altos de congestión vehicular junto con los valores más bajos de PIB per cápita, en comparación con Lima y Buenos Aires. Este posicionamiento sugiere una posible relación estructural entre limitaciones en la movilidad urbana y menores niveles de productividad económica relativa.

Lima y Buenos Aires, aunque también enfrentan problemas de congestión, muestran un desempeño económico per cápita relativamente superior, lo que podría estar asociado a diferencias en infraestructura, estructura productiva, densidad urbana o políticas de transporte. Estos contrastes refuerzan la importancia de analizar la congestión no solo como un fenómeno aislado, sino como parte de un sistema urbano más amplio.

Asimismo, se identifican ciertos valores atípicos que podrían estar influenciados por factores externos no capturados en el dataset, como shocks económicos, políticas públicas recientes o cambios en patrones de movilidad post-pandemia. Estos outliers representan oportunidades para profundizar el análisis mediante variables adicionales o fuentes complementarias.

Conclusiones y recomendaciones

A partir de los hallazgos observados, Bogotá emerge como la ciudad con la correlación más preocupante entre alta congestión vehicular y bajos indicadores de productividad económica, lo que la posiciona como una ciudad prioritaria para inversión en infraestructura de transporte y mejoras en movilidad urbana. Si bien la correlación observada no implica causalidad directa, la consistencia del patrón sugiere la existencia de cuellos de botella estructurales que podrían estar limitando su desempeño económico.

Se recomienda, en primer lugar, priorizar inversiones focalizadas en transporte público, gestión inteligente del tráfico y reducción de tiempos de traslado en ciudades con perfiles similares al de Bogotá. En segundo lugar, es fundamental validar la estabilidad y precisión de los indicadores de congestión mediante la incorporación de series de tiempo más largas y fuentes alternativas, lo que permitiría evaluar tendencias de largo plazo.

Finalmente, se sugiere ampliar el alcance del análisis incorporando variables adicionales como crecimiento económico, informalidad laboral, uso modal del transporte, densidad poblacional y emisiones, con el fin de construir un modelo más robusto que permita avanzar hacia conclusiones causales. Este enfoque fortalecerá la toma de decisiones estratégicas y permitirá maximizar el impacto económico y social de futuras inversiones en movilidad urbana.
